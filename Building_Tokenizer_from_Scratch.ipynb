```python
class BhojpuriTokenizerV1:
    def __init__(self, vocab):
        self.str_to_int = vocab
        self.int_to_str = { i:s for s,i in vocab.items()}

    def encode(self, text):
        preprocessed = re.split(r'([,.:;?_!"()\']|--|\s)', text)
        preprocessed = [item.strip() for item in preprocessed if item.strip()]

        preprocessed = [item if item in self.str_to_int
                        else "<|unk|>" for item in preprocessed]
        ids = [self.str_to_int[s] for s in preprocessed]
        return ids

    def decode(self, ids):
      text = " ".join([self.int_to_str[i] for i in ids])
      text = re.sub(r'\s+([,.:;?!"()\'])', r'\1', text)
      return text
```

# **Byte Pair Encoding (BPE) Implementations**

Byte Pair Encoding (BPE) is a data-driven subword tokenization method that builds a vocabulary of subword units by iteratively merging the most frequent pair of symbols. It bridges the gap between character-level and word-level models

$$
\text{Tokenize}(w) = s \setminus \{</w>\}
$$

```python
import re
import unicodedata
from collections import defaultdict

class BytePairEncoder:
    def __init__(self, vocab_size=None, num_merges=10000):
        self.vocab_size = vocab_size
        self.num_merges = num_merges
        self.bpe_codes = []

    @staticmethod
    def normalize_text(text):
        text = unicodedata.normalize('NFC', text)
        text = re.sub(r'[\u0964\u0965]', ' ', text)
        text = re.sub(r'[^\u0900-\u097F\s]', '', text)
        return text

    @staticmethod
    def tokenize_sentence(sentence):
        return [w for w in sentence.strip().split() if w]

    def build_corpus(self, texts):
        words = []
        for text in texts:
            norm = self.normalize_text(text)
            words.extend(self.tokenize_sentence(norm))
        return words

    def get_vocab(self, words):
        vocab = defaultdict(int)
        for word in words:
            symbols = tuple(word) + ('</w>',)
            vocab[symbols] += 1
        return vocab

    def get_stats(self, vocab):
        pairs = defaultdict(int)
        for symbols, freq in vocab.items():
            for i in range(len(symbols) - 1):
                pair = (symbols[i], symbols[i+1])
                pairs[pair] += freq
        return pairs

    def merge_vocab(self, pair, vocab):
        merged_vocab = {}
        bigram = re.escape(' '.join(pair))
        pattern = re.compile(r'(?<!\S)' + bigram + r'(?!\S)')
        new_symbol = ''.join(pair)
        for symbols, freq in vocab.items():
            word = ' '.join(symbols)
            new_word = pattern.sub(new_symbol, word)
            new_symbols = tuple(new_word.split(' '))
            merged_vocab[new_symbols] = freq
        return merged_vocab

    def learn_bpe(self, texts):
        words = self.build_corpus(texts)
        vocab = self.get_vocab(words)

        for i in range(self.num_merges):
            pairs = self.get_stats(vocab)
            if not pairs:
                break
            best_pair, freq = max(pairs.items(), key=lambda x: x[1])
            self.bpe_codes.append(best_pair)
            vocab = self.merge_vocab(best_pair, vocab)
            if self.vocab_size and len(vocab) >= self.vocab_size:
                break
        return self.bpe_codes

    def apply_bpe(self, word):
        symbols = list(word) + ['</w>']
        i = 0
        while i < len(symbols) - 1:
            pair = (symbols[i], symbols[i+1])
            if pair in self.bpe_codes:
                symbols[i:i+2] = [''.join(pair)]
                i = max(i-1, 0)
            else:
                i += 1
        if symbols[-1] == '</w>':
            symbols = symbols[:-1]
        return symbols

    def tokenize(self, text):
        norm = self.normalize_text(text)
        tokens = []
        for word in self.tokenize_sentence(norm):
            tokens.extend(self.apply_bpe(word))
        return tokens
```

```python
import re
import unicodedata
from collections import defaultdict

class ByteLevelBPETokenizer:
    def __init__(self, vocab_size=None, num_merges=30000):
        self.vocab_size = vocab_size
        self.num_merges = num_merges
        self.bpe_codes = {}
        self.byte_encoder = self.bytes_to_unicode()
        self.byte_decoder = {v: k for k, v in self.byte_encoder.items()}
        self.vocab = {}
        self._end_of_word = '</w>'

    @staticmethod
    def bytes_to_unicode():
        bs = list(range(ord('!'), ord('~') + 1)) + \
             list(range(ord(''), ord('') + 1)) + \
             list(range(ord(''), ord('') + 1))
        cs = bs[:]
        n = 0
        for b in range(256):
            if b not in bs:
                bs.append(b)
                cs.append(256 + n)
                n += 1
        return {b: chr(c) for b, c in zip(bs, cs)}

    @staticmethod
    def normalize_text(text):
        text = unicodedata.normalize('NFC', text)
        text = re.sub(r'[\u0964\u0965]', ' ', text)
        text = re.sub(r'[^\u0900-\u097F\s]', '', text)
        return text

    def _word_to_bpe_tokens(self, word):
        utf8_bytes = word.encode('utf-8')
        chars = [self.byte_encoder[b] for b in utf8_bytes]
        return tuple(chars + [self._end_of_word])

    def get_vocab_from_corpus(self, texts):
        vocab = defaultdict(int)
        for text in texts:
            normalized = self.normalize_text(text)
            for word in normalized.strip().split():
                symbol_seq = self._word_to_bpe_tokens(word)
                vocab[symbol_seq] += 1
        return vocab

    @staticmethod
    def get_pair_frequencies(vocab):
        pairs = defaultdict(int)
        for symbols, freq in vocab.items():
            for i in range(len(symbols) - 1):
                pairs[(symbols[i], symbols[i + 1])] += freq
        return pairs

    @staticmethod
    def merge_pair(pair, symbol_seq):
        a, b = pair
        merged = a + b
        seq = list(symbol_seq)
        i = 0
        new_seq = []
        while i < len(seq):
            if i < len(seq) - 1 and seq[i] == a and seq[i + 1] == b:
                new_seq.append(merged)
                i += 2
            else:
                new_seq.append(seq[i])
                i += 1
        return tuple(new_seq)

    def learn_bpe(self, texts):
        vocab = self.get_vocab_from_corpus(texts)
        for i in range(self.num_merges):
            pair_freqs = self.get_pair_frequencies(vocab)
            if not pair_freqs:
                break
            best_pair, best_freq = max(pair_freqs.items(), key=lambda kv: kv[1])
            if best_freq < 1:
                break
            self.bpe_codes[best_pair] = i
            new_vocab = {}
            for seq, freq in vocab.items():
                merged_seq = self.merge_pair(best_pair, seq)
                new_vocab[merged_seq] = freq
            vocab = new_vocab
            if self.vocab_size and len(vocab) >= self.vocab_size:
                break

        all_byte_chars = set(self.byte_encoder.values())
        next_id = 0
        for bch in sorted(all_byte_chars):
            self.vocab[bch] = next_id
            next_id += 1

        unique_tokens = set()
        for seq in vocab:
            unique_tokens.update(seq)

        merged_tokens = sorted(unique_tokens.difference(all_byte_chars))
        for tok in merged_tokens:
            self.vocab[tok] = next_id
            next_id += 1

    def bpe_on_symbol_seq(self, symbol_seq):
        seq = list(symbol_seq)
        i = 0
        while i < len(seq) - 1:
            pair = (seq[i], seq[i + 1])
            if pair in self.bpe_codes:
                merged = seq[i] + seq[i + 1]
                seq[i:i + 2] = [merged]
                i = max(i - 1, 0)
            else:
                i += 1
        if seq[-1] == self._end_of_word:
            seq = seq[:-1]
        return seq

    def encode(self, text):
        normalized = self.normalize_text(text)
        tokens = []
        for word in normalized.strip().split():
            symbol_seq = self._word_to_bpe_tokens(word)
            bpe_tokens = self.bpe_on_symbol_seq(symbol_seq)
            for btok in bpe_tokens:
                if btok in self.vocab:
                    tokens.append(self.vocab[btok])
                else:
                    for ch in btok:
                        tokens.append(self.vocab[ch])
        return tokens

    def decode(self, token_ids):
        pieces = [list(self.vocab.keys())[list(self.vocab.values()).index(tid)] for tid in token_ids]
        byte_list = []
        for piece in pieces:
            for ch in piece:
                b = self.byte_decoder[ch]
                byte_list.append(b)
        text_bytes = bytes(byte_list)
        return text_bytes.decode('utf-8', errors='ignore')
```

```python
import json
import csv

with open("/content/huggingface_data/dataset.json", "r", encoding="utf-8") as f:
    data = json.load(f)


# Extract Bhojpuri text from each row and write to CSV
save_dir = "huggingface_data"
os.makedirs(save_dir, exist_ok=True)
csv_path = os.path.join(save_dir, "bhojpuri_texts.csv")

with open(csv_path, "w", newline="", encoding="utf-8") as fout:
    writer = csv.writer(fout)
    writer.writerow(["text"])  # header
    for entry in data["rows"]:
        row_dict = entry["row"]
        # The text field is the third feature: find its key
        # In this dataset, the third feature name is the long Bhojpuri sentence in the "features" list
        text_key = data["features"][2]["name"]
        text_value = row_dict.get(text_key, "").strip()
        if text_value:
            writer.writerow([text_value])

print(f"Saved CSV with only text to: {csv_path}")
```
